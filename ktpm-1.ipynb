# 1. Chương trình chứa lỗi

```python
def xep_loai(gpa):

    # S1
    if gpa >= 3.6:
        return "Xuất sắc"

    # S2
    elif gpa >= 3.2:
        return "Giỏi"

    # S3
    elif gpa >= 2.5:
        return "Khá"

    # S4 (FAULT)
    elif gpa >= 2.0:
        return "Khá"

    # S5
    else:
        return "Yếu"
```

## 2. Chương trình đúng sau khi sửa lỗi

```python
def xep_loai(gpa):

    if gpa >= 3.6:
        return "Xuất sắc"

    elif gpa >= 3.2:
        return "Giỏi"

    elif gpa >= 2.5:
        return "Khá"

    elif gpa >= 2.0:
        return "Trung bình"

    else:
        return "Yếu"
```

## 3. Bộ Test case

```python
tests = [
    ("T1", 3.8, "Xuất sắc"),
    ("T2", 3.4, "Giỏi"),
    ("T3", 2.9, "Khá"),
    ("T4", 2.2, "Trung bình"),
    ("T5", 2.1, "Trung bình"),
    ("T6", 1.8, "Yếu")
]
```

## 4. Chạy kiểm thử

```python
results = []

for name, gpa, expected in tests:

    output = xep_loai(gpa)

    if output == expected:
        status = "PASS"
    else:
        status = "FAIL"

    results.append([name, gpa, expected, output, status])

for row in results:
    print(row)
```

## 7. Tính Tarantula

```python
import math

total_failed = 2
total_passed = 4

data = {
    "S1":(2,4),
    "S2":(2,3),
    "S3":(2,2),
    "S4":(2,0),
    "S5":(0,1)
}

for stmt,(failed,passed) in data.items():

    if failed == 0 and passed == 0:
        score = 0
    else:
        score = (failed/total_failed)/(
            (failed/total_failed)+
            (passed/total_passed)
        )

    print(stmt, round(score,3))
```

## 8. Tính Ochiai

```python
for stmt,(failed,passed) in data.items():

    if failed == 0:
        score = 0
    else:
        score = failed / math.sqrt(
            total_failed*(failed+passed)
        )

    print(stmt, round(score,3))
```
